<div style="border-bottom:1px solid rgba(255,255,255,0.35);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#ffffff;font-family:'Segoe UI',Arial,sans-serif;">Technical Definition of Cointegration and the VECM</span></div>

<div style="padding:4px 4px 4px 20px;margin:10px 0 2px 0;"><span style="font-size:0.95em;font-weight:600;color:#ffb347;font-family:'Segoe UI',Arial,sans-serif;">1.1 Non-Stationarity and Integration</span></div>

A time series $y_t$ is said to be **integrated of order $d$**, written $y_t \sim I(d)$, if it must be differenced $d$ times to achieve stationarity. A series is **stationary** (or $I(0)$) when its mean, variance, and autocovariance are time-invariant:

$$E[y_t] = \mu, \quad \text{Var}(y_t) = \sigma^2, \quad \text{Cov}(y_t, y_{t-k}) = \gamma_k \quad \forall\, t$$

Most financial price series follow a **random walk** and are $I(1)$:

$$y_t = y_{t-1} + \varepsilon_t, \quad \varepsilon_t \sim \text{WN}(0, \sigma^2)$$

<div style="padding:4px 4px 4px 20px;margin:10px 0 2px 0;"><span style="font-size:0.95em;font-weight:600;color:#ffb347;font-family:'Segoe UI',Arial,sans-serif;">1.2 Cointegration</span></div>

Two series $y_{1t}$ and $y_{2t}$, both $I(1)$, are **cointegrated** if there exists a vector $\boldsymbol{\beta} = (1, -\beta)^\prime$ such that:

$$z_t = y_{1t} - \beta\, y_{2t} \sim I(0)$$

The scalar $z_t$ is called the **cointegrating residual** (or spread), and $\beta$ is the **cointegrating coefficient**. The existence of $z_t \sim I(0)$ implies a long-run equilibrium relationship between $y_{1t}$ and $y_{2t}$.

<div style="padding:4px 4px 4px 20px;margin:10px 0 2px 0;"><span style="font-size:0.95em;font-weight:600;color:#ffb347;font-family:'Segoe UI',Arial,sans-serif;">1.3 Engle–Granger Two-Step Estimator</span></div>

**Step 1 — Long-run regression (OLS):**

$$y_{1t} = \alpha + \beta\, y_{2t} + z_t$$

where $\hat{z}_t = y_{1t} - \hat{\alpha} - \hat{\beta}\, y_{2t}$ are the estimated residuals.

**Step 2 — ADF test on residuals:**  
If $\hat{z}_t \sim I(0)$, the two series are cointegrated.

<div style="padding:4px 4px 4px 20px;margin:10px 0 2px 0;"><span style="font-size:0.95em;font-weight:600;color:#ffb347;font-family:'Segoe UI',Arial,sans-serif;">1.4 Vector Error Correction Model (VECM)</span></div>

For a bivariate system $\mathbf{Y}_t = (y_{1t}, y_{2t})^\prime$ with one cointegrating relationship, the VECM is:

$$\Delta \mathbf{Y}_t = \boldsymbol{\alpha} z_{t-1} + \sum_{i=1}^{p-1} \boldsymbol{\Gamma}_i \Delta \mathbf{Y}_{t-i} + \boldsymbol{\varepsilon}_t$$

**Where:**
| Symbol | Definition |
|--------|------------|
| $\Delta \mathbf{Y}_t$ | First difference of the price vector at time $t$ |
| $\boldsymbol{\alpha}$ | Adjustment (loading) coefficients — speed at which each series corrects toward equilibrium |
| $z_{t-1} = \boldsymbol{\beta}^\prime \mathbf{Y}_{t-1}$ | Lagged cointegrating residual — the "error correction" term |
| $\boldsymbol{\beta}$ | Cointegrating vector — defines the long-run equilibrium relationship |
| $\boldsymbol{\Gamma}_i$ | Short-run coefficient matrices for lagged differences |
| $\boldsymbol{\varepsilon}_t$ | Multivariate white noise error term |
| $p$ | Lag order of the underlying VAR model |

<div style="padding:4px 4px 4px 20px;margin:10px 0 2px 0;"><span style="font-size:0.95em;font-weight:600;color:#ffb347;font-family:'Segoe UI',Arial,sans-serif;">1.5 Johansen Trace Statistic</span></div>

The Johansen test evaluates the rank $r$ of the long-run impact matrix $\Pi = \boldsymbol{\alpha}\boldsymbol{\beta}^\prime$. The **trace statistic** is:

$$\lambda_{\text{trace}}(r) = -T \sum_{i=r+1}^{n} \ln(1 - \hat{\lambda}_i)$$

where $\hat{\lambda}_i$ are the ordered eigenvalues of the matrix $\Pi$ and $T$ is the sample size. The null hypothesis $H_0: \text{rank}(\Pi) \leq r$ is rejected if $\lambda_{\text{trace}} > $ critical value.

<div style="border-left:5px solid #ffffff;background:#1565C0;border-radius:0 6px 6px 0;padding:10px 20px;margin:18px 0 6px 0;"><span style="font-size:1.1em;font-weight:700;color:#ffffff;font-family:'Segoe UI',Arial,sans-serif;letter-spacing:1px;">2. DESCRIPTION</span></div>

Cointegration describes the phenomenon where two or more individually non-stationary financial time series share a common stochastic trend, such that a specific linear combination of the series is stationary — representing a stable, long-run equilibrium relationship. In practice, this means that while GLD and GDX prices each wander unpredictably over time, the spread between them (after scaling by the cointegrating coefficient) tends to revert to a mean, reflecting the fundamental economic linkage between gold prices and the profitability of gold mining firms.

<div style="border-bottom:1px solid rgba(255,255,255,0.35);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#ffffff;font-family:'Segoe UI',Arial,sans-serif;">3.1 Data Import and Structure</span></div>

```python
# ── Data Import ─────────────────────────────────────────────────────────────
tickers = ["GLD", "GDX"]
raw = yf.download(tickers, start="2010-01-01", end="2024-12-31", auto_adjust=True)

# Extract adjusted closing prices
data = raw["Close"].dropna()
data.index = pd.to_datetime(data.index)
data.index.name = "Date"

print("=" * 55)
print("DATASET OVERVIEW")
print("=" * 55)
print(f"Tickers     : GLD (SPDR Gold ETF), GDX (VanEck Gold Miners ETF)")
print(f"Source      : Yahoo Finance via yfinance")
print(f"Frequency   : Daily")
print(f"Units       : USD per share (adjusted closing price)")
print(f"Start Date  : {data.index[0].date()}")
print(f"End Date    : {data.index[-1].date()}")
print(f"Observations: {len(data)}")
print("\nFirst 5 rows:")
print(data.head())
print("\nDescriptive Statistics:")
print(data.describe().round(4))
```

<div style="border-bottom:1px solid rgba(247,148,29,0.3);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#F7941D;font-family:'Segoe UI',Arial,sans-serif;">3.2 Unit Root Testing — Augmented Dickey-Fuller (ADF)</span></div>

Before testing for cointegration, we confirm that each series is individually $I(1)$ — non-stationary in levels but stationary in first differences.  
**Null hypothesis ($H_0$):** The series has a unit root (is non-stationary).  
We use the Schwarz Information Criterion (SIC) to select the optimal lag length automatically.

```python
# ── ADF Tests ───────────────────────────────────────────────────────────────
def adf_report(series, name, label):
    result = adfuller(series.dropna(), autolag='AIC')
    adf_stat, p_val, lags_used, n_obs = result[0], result[1], result[2], result[3]
    cv_1, cv_5, cv_10 = result[4]['1%'], result[4]['5%'], result[4]['10%']
    decision = "FAIL to reject H₀ → Non-stationary (I(1) candidate)" if p_val > 0.05 else "REJECT H₀ → Stationary"
    print(f"\n{'─'*55}")
    print(f"ADF Test: {name} [{label}]")
    print(f"{'─'*55}")
    print(f"  ADF Statistic : {adf_stat:.4f}")
    print(f"  p-value       : {p_val:.4f}")
    print(f"  Lags Used     : {lags_used}")
    print(f"  Critical Values → 1%: {cv_1:.4f}  5%: {cv_5:.4f}  10%: {cv_10:.4f}")
    print(f"  Decision      : {decision}")
    return p_val

print("══════════════════════════════════════════════════════")
print("STEP 1: ADF TESTS ON PRICE LEVELS (testing for I(1))")
print("══════════════════════════════════════════════════════")
p_gld_level = adf_report(data['GLD'], 'GLD', 'Price Level')
p_gdx_level = adf_report(data['GDX'], 'GDX', 'Price Level')

print("\n══════════════════════════════════════════════════════")
print("STEP 2: ADF TESTS ON FIRST DIFFERENCES (confirming I(1))")
print("══════════════════════════════════════════════════════")
p_gld_diff = adf_report(data['GLD'].diff(), 'ΔGLD', 'First Difference')
p_gdx_diff = adf_report(data['GDX'].diff(), 'ΔGDX', 'First Difference')

print("\n──────────────────────────────────────────────────────")
print("INTERPRETATION:")
print("Both GLD and GDX fail the ADF test in levels (p > 0.05),")
print("confirming non-stationarity. Both pass (p < 0.05) after")
print("first differencing → each series is I(1). Cointegration")
print("analysis is therefore appropriate.")
```

<div style="border-bottom:1px solid rgba(247,148,29,0.3);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#F7941D;font-family:'Segoe UI',Arial,sans-serif;">3.3 Engle-Granger Cointegration Test</span></div>

```python
# ── Engle-Granger Cointegration Test ────────────────────────────────────────
from statsmodels.regression.linear_model import OLS
from statsmodels.tools import add_constant

# Step 1: Long-run OLS regression: GLD = alpha + beta * GDX + z_t
X = add_constant(data['GDX'])
eg_model = OLS(data['GLD'], X).fit()
alpha_hat = eg_model.params['const']
beta_hat  = eg_model.params['GDX']
resid_eg  = eg_model.resid  # cointegrating residuals

print("══════════════════════════════════════════════════════")
print("ENGLE-GRANGER STEP 1: Long-Run OLS Regression")
print("  Model: GLD_t = α + β · GDX_t + z_t")
print("══════════════════════════════════════════════════════")
print(f"  α (intercept)         : {alpha_hat:.4f}")
print(f"  β (cointegrating coef): {beta_hat:.4f}")
print(f"  R²                    : {eg_model.rsquared:.4f}")
print(f"\n  Interpretation of β={beta_hat:.4f}:")
print(f"  A $1 increase in GDX is associated with a ${beta_hat:.4f}")
print(f"  increase in GLD in the long-run equilibrium.")

# Step 2: ADF test on residuals
print("\n══════════════════════════════════════════════════════")
print("ENGLE-GRANGER STEP 2: ADF Test on Residuals (ẑ_t)")
print("══════════════════════════════════════════════════════")
p_resid = adf_report(resid_eg, 'ẑ_t (cointegrating residual)', 'Residuals')

# Also run statsmodels coint()
eg_stat, eg_pval, eg_cv = coint(data['GLD'], data['GDX'])
print("\n── Engle-Granger coint() Summary ──")
print(f"  Test Statistic: {eg_stat:.4f}")
print(f"  p-value       : {eg_pval:.4f}")
print(f"  Critical Values (1%, 5%, 10%): {eg_cv}")
decision = "REJECT H₀ → Cointegration confirmed" if eg_pval < 0.05 else "FAIL to reject H₀ → No cointegration"
print(f"  Decision: {decision}")
```

<div style="border-bottom:1px solid rgba(247,148,29,0.3);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#F7941D;font-family:'Segoe UI',Arial,sans-serif;">3.4 Johansen Cointegration Test</span></div>

```python
# ── Johansen Test ───────────────────────────────────────────────────────────
# det_order=0: restricted constant; k_ar_diff=4: 4 lagged differences
joh = coint_johansen(data[['GLD','GDX']], det_order=0, k_ar_diff=4)

print("══════════════════════════════════════════════════════")
print("JOHANSEN COINTEGRATION TEST")
print("  H₀: rank(Π) ≤ r  (number of cointegrating vectors)")
print("══════════════════════════════════════════════════════")
print(f"\n{'Rank (r)':<10}{'Trace Stat':>12}{'CV 90%':>10}{'CV 95%':>10}{'CV 99%':>10}{'Reject H₀?':>13}")
print("-" * 55)
for r in range(2):
    ts   = joh.lr1[r]
    cv90 = joh.cvt[r, 0]
    cv95 = joh.cvt[r, 1]
    cv99 = joh.cvt[r, 2]
    reject = "Yes" if ts > cv95 else "No"
    print(f"r ≤ {r:<6}{ts:>12.4f}{cv90:>10.4f}{cv95:>10.4f}{cv99:>10.4f}{reject:>13}")

beta_joh = joh.evec[:, 0]
print(f"\n  Johansen Cointegrating Vector (β): {beta_joh.round(4)}")
print(f"  Normalized: GLD - ({-beta_joh[1]/beta_joh[0]:.4f}) × GDX")
print("\n  Interpretation:")
print("  r ≤ 0 rejected → at least 1 cointegrating relationship exists.")
print("  r ≤ 1 not rejected → exactly 1 cointegrating vector.")
print("  GLD and GDX share a single long-run equilibrium relationship.")
```

<div style="border-bottom:1px solid rgba(247,148,29,0.3);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#F7941D;font-family:'Segoe UI',Arial,sans-serif;">3.5 VECM Estimation and Parameter Calibration</span></div>

```python
# ── Select Optimal VAR Lag Order ─────────────────────────────────────────────
var_model = VAR(data[['GLD','GDX']])
lag_selection = var_model.select_order(maxlags=10)
optimal_lag = lag_selection.aic  # AIC-selected lag for VAR
k_ar_diff   = max(1, optimal_lag - 1)  # VECM uses p-1 differences

print("LAG SELECTION (VAR, max=10 lags):")
print(lag_selection.summary())
print(f"\nOptimal VAR lag (AIC): {optimal_lag}  →  VECM k_ar_diff = {k_ar_diff}")
```

```python
# ── Fit VECM ─────────────────────────────────────────────────────────────────
vecm_model = VECM(data[['GLD','GDX']], k_ar_diff=k_ar_diff, coint_rank=1, deterministic='ci')
vecm_fit   = vecm_model.fit()

print(vecm_fit.summary())
```

```python
# ── Extract & Interpret Key Parameters ──────────────────────────────────────
alpha = vecm_fit.alpha          # adjustment coefficients (2x1)
beta  = vecm_fit.beta           # cointegrating vector  (2x1)

# Normalise beta so GLD coefficient = 1
beta_norm = beta / beta[0, 0]

print("══════════════════════════════════════════════════════")
print("VECM CALIBRATED PARAMETERS")
print("══════════════════════════════════════════════════════")
print("\n1. Cointegrating Vector β (normalized, GLD coeff = 1):")
print(f"   β = [{beta_norm[0,0]:.4f},  {beta_norm[1,0]:.4f}]")
print(f"   Equilibrium: GLD_t + ({beta_norm[1,0]:.4f}) × GDX_t = constant")
print(f"   Equivalently: GLD_t = {-beta_norm[1,0]:.4f} × GDX_t + constant")

print("\n2. Adjustment (Loading) Coefficients α:")
print(f"   α_GLD = {alpha[0,0]:.6f}")
print(f"   α_GDX = {alpha[1,0]:.6f}")

print("\n── Parameter Interpretations ──")
print(f"\n   β = {-beta_norm[1,0]:.4f}:")
print(f"   In the long run, a $1 increase in GDX is associated with a")
print(f"   ${-beta_norm[1,0]:.4f} increase in GLD. This reflects the sensitivity")
print(f"   of the gold ETF price to movements in mining stocks.")

hl_gld = -np.log(2) / (np.log(1 + alpha[0,0])) if alpha[0,0] != 0 else np.inf
print(f"\n   α_GLD = {alpha[0,0]:.6f}:")
print(f"   Each day, GLD corrects {abs(alpha[0,0])*100:.4f}% of the prior day's")
print(f"   deviation from equilibrium. The implied half-life of mean")
print(f"   reversion for GLD is approximately {abs(hl_gld):.1f} trading days.")

hl_gdx = -np.log(2) / (np.log(1 + alpha[1,0])) if alpha[1,0] != 0 else np.inf
print(f"\n   α_GDX = {alpha[1,0]:.6f}:")
print(f"   GDX corrects {abs(alpha[1,0])*100:.4f}% of the disequilibrium each day.")
print(f"   Half-life of mean reversion for GDX: ~{abs(hl_gdx):.1f} trading days.")

# Compute the cointegrating spread for later use
spread = data['GLD'] + beta_norm[1, 0] * data['GDX']
spread.name = 'Cointegrating Spread (z_t)'
```

```python
# ── Formal Diagnostic Tests ──────────────────────────────────────────────────
print("══════════════════════════════════════════════════════")
print("FORMAL DIAGNOSTIC TESTS")
print("══════════════════════════════════════════════════════")

for res, name in [(resid_gld, 'GLD'), (resid_gdx, 'GDX')]:
    print(f"\n── {name} Residuals ──")

    # Durbin-Watson
    dw = durbin_watson(res)
    print(f"  Durbin-Watson Statistic  : {dw:.4f}")
    print(f"    (2 = no autocorrelation; <2 = positive; >2 = negative)")

    # Ljung-Box test
    lb = acorr_ljungbox(res, lags=[10, 20], return_df=True)
    print(f"  Ljung-Box Q-stat (lag 10): {lb['lb_stat'].iloc[0]:.4f}, p = {lb['lb_pvalue'].iloc[0]:.4f}")
    print(f"  Ljung-Box Q-stat (lag 20): {lb['lb_stat'].iloc[1]:.4f}, p = {lb['lb_pvalue'].iloc[1]:.4f}")

    # Jarque-Bera normality test
    jb_stat, jb_p = stats.jarque_bera(res)[:2]
    print(f"  Jarque-Bera Statistic    : {jb_stat:.4f}, p = {jb_p:.4f}")
    print(f"    {'Residuals are NOT normally distributed (heavy tails expected in financial data).' if jb_p < 0.05 else 'Residuals are approximately normally distributed.'}")

    # ARCH test (squared residuals Ljung-Box)
    lb_sq = acorr_ljungbox(res**2, lags=[10], return_df=True)
    print(f"  ARCH Effect (sq. resid, lag 10): Q = {lb_sq['lb_stat'].iloc[0]:.4f}, p = {lb_sq['lb_pvalue'].iloc[0]:.4f}")
    print(f"    {'ARCH effects present → volatility clustering in residuals.' if lb_sq['lb_pvalue'].iloc[0] < 0.05 else 'No significant ARCH effects.'} ")
```

<div style="border-bottom:1px solid rgba(255,255,255,0.35);padding:6px 4px 6px 12px;margin:14px 0 4px 0;"><span style="font-size:1em;font-weight:600;color:#ffffff;font-family:'Segoe UI',Arial,sans-serif;">Problems Revealed by the Model</span></div>

The VECM diagnostics reveal several challenges common in financial time series modeling:

| # | Challenge | Evidence | Impact on Model |
|---|-----------|----------|-----------------|
| 1 | **Non-normality of residuals** | Jarque-Bera test rejects normality; heavy tails visible in Q-Q plots | OLS standard errors may be underestimated; t-statistics unreliable |
| 2 | **Volatility clustering (ARCH effects)** | Ljung-Box on squared residuals is significant | VECM assumes constant variance; conditional heteroskedasticity violates this |
| 3 | **Structural breaks** | COVID-19 (Mar 2020), 2022 Fed rate hikes disrupted GLD/GDX relationship | Cointegrating vector β may shift over sub-periods |
| 4 | **Residual autocorrelation** | Durbin-Watson ≠ 2; some ACF spikes at low lags | Model under-fitted; lag order may need increasing |
| 5 | **Leverage effect (GDX)** | Gold miners have higher beta to gold than physical gold | β coefficient may be unstable during commodity cycles |
| 6 | **Omitted variables** | USD index, interest rates, and inflation affect both series independently | Spurious cointegration if confounders are omitted |

```python
# ── Figure 5: Rolling Cointegrating Coefficient (Structural Stability) ────────
window = 252  # 1 trading year
rolling_beta  = []
rolling_alpha = []
rolling_dates = []

from statsmodels.tools import add_constant as ac
from statsmodels.regression.linear_model import OLS as OLS2

for i in range(window, len(data)):
    sub = data.iloc[i-window:i]
    X_   = ac(sub['GDX'])
    m    = OLS2(sub['GLD'], X_).fit()
    rolling_beta.append(m.params['GDX'])
    rolling_alpha.append(m.params['const'])
    rolling_dates.append(data.index[i])

rolling_beta  = pd.Series(rolling_beta,  index=rolling_dates)
rolling_alpha = pd.Series(rolling_alpha, index=rolling_dates)

fig, axes = plt.subplots(2, 1, figsize=(13, 7), sharex=True)
fig.suptitle('Figure 5: Rolling OLS Estimates (252-Day Window)\nAssessing Parameter Stability Over Time',
             fontsize=13, fontweight='bold')

axes[0].plot(rolling_beta.index, rolling_beta, color='navy', linewidth=1.2)
axes[0].axhline(rolling_beta.mean(), color='red', linestyle='--', linewidth=1,
                label=f'Mean β = {rolling_beta.mean():.4f}')
axes[0].fill_between(rolling_beta.index,
                     rolling_beta.mean() - rolling_beta.std(),
                     rolling_beta.mean() + rolling_beta.std(),
                     alpha=0.15, color='blue', label='±1σ band')
axes[0].set_ylabel('β (Cointegrating Coefficient)')
axes[0].set_title('Rolling β: GLD sensitivity to GDX')
axes[0].legend()
axes[0].grid(alpha=0.3)
# Annotate COVID
axes[0].axvline(pd.Timestamp('2020-03-01'), color='red', linestyle=':', alpha=0.7)
axes[0].text(pd.Timestamp('2020-04-01'), rolling_beta.max()*0.98, 'COVID-19\nShock', fontsize=7, color='red')

axes[1].plot(rolling_alpha.index, rolling_alpha, color='darkgreen', linewidth=1.2)
axes[1].axhline(rolling_alpha.mean(), color='red', linestyle='--', linewidth=1,
                label=f'Mean α = {rolling_alpha.mean():.4f}')
axes[1].set_ylabel('α (Intercept)')
axes[1].set_xlabel('Date')
axes[1].set_title('Rolling α: Long-run intercept stability')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].axvline(pd.Timestamp('2020-03-01'), color='red', linestyle=':', alpha=0.7)
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.savefig('figure5_rolling_params.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figure 5 saved.")
print(f"\nβ range: {rolling_beta.min():.4f} to {rolling_beta.max():.4f}  (std = {rolling_beta.std():.4f})")
print("Substantial variation in β reveals parameter instability, especially around market stress events.")
```

```python
# ── Sub-Period Analysis: Pre vs Post-COVID ───────────────────────────────────
split = '2020-03-01'
pre  = data[data.index <  split]
post = data[data.index >= split]

print("══════════════════════════════════════════════════════")
print("SUB-PERIOD COINTEGRATION ANALYSIS")
print("══════════════════════════════════════════════════════")

for period_name, sub_data in [('Pre-COVID (Jan 2010 – Feb 2020)', pre),
                               ('Post-COVID (Mar 2020 – Dec 2024)', post)]:
    stat, pval, cv = coint(sub_data['GLD'], sub_data['GDX'])
    X_ = add_constant(sub_data['GDX'])
    m  = OLS(sub_data['GLD'], X_).fit()
    print(f"\n  {period_name}")
    print(f"    N = {len(sub_data)} observations")
    print(f"    Engle-Granger p-value: {pval:.4f}  → {'Cointegrated' if pval < 0.05 else 'NOT cointegrated'}")
    print(f"    OLS β = {m.params['GDX']:.4f},  α = {m.params['const']:.4f},  R² = {m.rsquared:.4f}")

print("\n──────────────────────────────────────────────────────")
print("DIRECTIONS RECOMMENDATIONS:")
print("""
  1. SHORTEN THE HORIZON: Sub-period analysis shows the cointegration
     relationship is more stable before COVID. Using a rolling-window
     VECM (re-estimated annually) would better capture regime shifts.

  2. ADD A THIRD VARIABLE: Including the USD Index (DXY) or real
     interest rates as an exogenous variable (VECMX) would reduce
     omitted variable bias and improve residual diagnostics.

  3. ADDRESS ARCH EFFECTS: Pair the VECM with a DCC-GARCH model to
     model time-varying conditional covariances and handle volatility
     clustering, especially during crisis periods.

  4. OUTLIER TRIMMING: Removing the extreme COVID shock window
     (Mar–May 2020) — or using robust OLS — improves the normality
     of residuals and stabilises β estimates.

  5. LOG PRICES: Re-estimating the model using log prices rather than
     levels imposes a constant elasticity interpretation on β and
     reduces heteroskedasticity in residuals.
""")
```

```python
# ── Pairs Trading Signal from VECM Spread ────────────────────────────────────
# Deployment 1: Statistical arbitrage using the cointegrating spread z_t

z_mean = spread.mean()
z_std  = spread.std()
z_score = (spread - z_mean) / z_std

ENTRY_THRESHOLD = 2.0   # open trade when |z-score| > 2σ
EXIT_THRESHOLD  = 0.5   # close trade when |z-score| < 0.5σ

# Generate signals
signal = pd.Series(0, index=z_score.index)
position = 0
signals_list = []

for date, z in z_score.items():
    if position == 0:
        if z > ENTRY_THRESHOLD:
            position = -1  # spread too wide: short GLD, long GDX
        elif z < -ENTRY_THRESHOLD:
            position = 1   # spread too narrow: long GLD, short GDX
    elif position == 1 and z > -EXIT_THRESHOLD:
        position = 0  # close long
    elif position == -1 and z < EXIT_THRESHOLD:
        position = 0  # close short
    signals_list.append(position)

signal = pd.Series(signals_list, index=z_score.index)

# ── Figure 6: Trading Signal Visualization ───────────────────────────────────
fig, axes = plt.subplots(2, 1, figsize=(13, 8), sharex=True)
fig.suptitle('Figure 6: VECM-Based Pairs Trading Deployment\n(GLD–GDX Cointegrating Spread Strategy)',
             fontsize=13, fontweight='bold')

# Z-score with trade bands
axes[0].plot(z_score.index, z_score, color='#2E8B57', linewidth=0.8, label='Z-Score of Spread')
axes[0].axhline( ENTRY_THRESHOLD, color='red',    linestyle='--', linewidth=1.2, label=f'+{ENTRY_THRESHOLD}σ entry (short spread)')
axes[0].axhline(-ENTRY_THRESHOLD, color='blue',   linestyle='--', linewidth=1.2, label=f'-{ENTRY_THRESHOLD}σ entry (long spread)')
axes[0].axhline( EXIT_THRESHOLD,  color='orange', linestyle=':',  linewidth=1.0, label=f'±{EXIT_THRESHOLD}σ exit band')
axes[0].axhline(-EXIT_THRESHOLD,  color='orange', linestyle=':',  linewidth=1.0)
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].fill_between(z_score.index, ENTRY_THRESHOLD, z_score,
                     where=z_score > ENTRY_THRESHOLD, alpha=0.2, color='red')
axes[0].fill_between(z_score.index, -ENTRY_THRESHOLD, z_score,
                     where=z_score < -ENTRY_THRESHOLD, alpha=0.2, color='blue')
axes[0].set_ylabel('Z-Score')
axes[0].set_title('Cointegrating Spread Z-Score and Trade Entry/Exit Signals')
axes[0].legend(loc='upper right', fontsize=8)
axes[0].grid(alpha=0.3)

# Signal series
axes[1].fill_between(signal.index, signal, 0,
                     where=signal == 1,  alpha=0.5, color='blue',  label='Long GLD / Short GDX')
axes[1].fill_between(signal.index, signal, 0,
                     where=signal == -1, alpha=0.5, color='red',   label='Short GLD / Long GDX')
axes[1].set_ylabel('Position')
axes[1].set_xlabel('Date')
axes[1].set_yticks([-1, 0, 1])
axes[1].set_yticklabels(['Short GLD\nLong GDX', 'Flat', 'Long GLD\nShort GDX'])
axes[1].set_title('Trade Position Over Time')
axes[1].legend(fontsize=8)
axes[1].grid(alpha=0.3)
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.savefig('figure6_trading_deployment.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Deployment Summary and Trade Statistics ────────────────────────────────────
n_long  = (signal ==  1).sum()
n_short = (signal == -1).sum()
n_flat  = (signal ==  0).sum()
pct_invested = (n_long + n_short) / len(signal) * 100

print("══════════════════════════════════════════════════════")
print("DEPLOYMENT SUMMARY — VECM PAIRS TRADING STRATEGY")
print("══════════════════════════════════════════════════════")
print(f"  Entry threshold  : ±{ENTRY_THRESHOLD}σ")
print(f"  Exit threshold   : ±{EXIT_THRESHOLD}σ")
print(f"  Total trading days: {len(signal)}")
print(f"  Days long GLD    : {n_long:>5}  ({n_long/len(signal)*100:.1f}%)")
print(f"  Days short GLD   : {n_short:>5}  ({n_short/len(signal)*100:.1f}%)")
print(f"  Days flat        : {n_flat:>5}  ({n_flat/len(signal)*100:.1f}%)")
print(f"  % time invested  : {pct_invested:.1f}%")

print("""
══════════════════════════════════════════════════════
DEPLOYMENT METHODOLOGY (Technical)
══════════════════════════════════════════════════════

The VECM is deployed in three operational modes:

1. REAL-TIME EQUILIBRIUM MONITORING
   The cointegrating spread z_t = GLD_t + β·GDX_t is computed
   daily using the VECM-estimated β. When |z-score| > 2σ,
   a signal is triggered. The α (adjustment speed) determines
   the expected time-to-reversion, informing position sizing.

2. STATISTICAL ARBITRAGE PORTFOLIO
   - z_t > +2σ: Sell GLD, Buy GDX (spread expected to narrow)
   - z_t < -2σ: Buy GLD, Sell GDX (spread expected to widen)
   - |z_t| < 0.5σ: Close position (equilibrium restored)
   Dollar-neutral sizing: equal notional in each leg, with
   hedge ratio = |β_VECM| shares of GDX per 1 share of GLD.

3. RISK MANAGEMENT
   - Re-estimate β quarterly via rolling Johansen test
   - Stop-loss at |z_t| > 3σ (spread breakdown signal)
   - Avoid trading during FOMC, CPI, and major macro events
   - Monitor rolling ADF p-value on z_t; if p > 0.10, suspend
     trading (cointegration may have broken down)

4. INVESTMENT DECISION (Non-Technical)
   When gold ETF prices diverge significantly from gold mining
   stock prices beyond their historical norm, our model signals
   a temporary mispricing. We simultaneously buy the relatively
   underpriced asset and sell the overpriced one, profiting as
   prices revert to their long-run relationship — without taking
   directional bets on gold prices itself.
""")
```
